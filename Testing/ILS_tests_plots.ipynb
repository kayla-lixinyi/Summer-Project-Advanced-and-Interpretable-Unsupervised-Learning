```python
a = np.random.normal(0, 0.5, (200, 2))
b = np.random.normal(3, 0.5, (200, 2))
c = np.random.normal(5.5, 0.5, (200, 2))
d = np.random.normal(6.8, 0.25, (200, 2))
e = np.random.normal(7.4, 0.125, (200, 2))
f = np.random.normal(7.75, 0.06725, (200, 2))
g = np.random.normal(7.91, 0.03225, (200, 2))
test = np.concatenate((a, b, c, d, e, f, g), axis = 0)
```

```python
testILS = ILS(min_cluster_size = 150).fit(test)
maxlst = [1 if i in testILS.maxima else 0 for i in range(len(testILS.rmin))]
plt.plot(maxlst)
testILS.coloured_rmin()
testILS.plot_labels()
```
