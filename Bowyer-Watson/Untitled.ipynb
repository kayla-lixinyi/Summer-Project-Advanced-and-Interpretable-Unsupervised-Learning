```python
import numpy as np
import matplotlib.pyplot as plt
from Bowyer_Watson import create_voronoi
import scipy.spatial.distance
```

```python
metric = getattr(scipy.spatial.distance, 'euclidean')
points = np.random.uniform(-2, 2, (20, 2))
graph = create_voronoi(points, metric)
```

```python
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])
for node in graph:
    node.plot_2D()
plt.show()
```
