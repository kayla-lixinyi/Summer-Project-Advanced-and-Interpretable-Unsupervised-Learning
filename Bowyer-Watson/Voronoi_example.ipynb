```python
import numpy as np
import matplotlib.pyplot as plt
from Bowyer_Watson import create_voronoi
import scipy.spatial.distance
```

```python
metric = getattr(scipy.spatial.distance, 'euclidean')
points = np.random.uniform(-2, 2, (20, 2))
graph = create_voronoi(points, metric)
```

```python
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])
plt.scatter(points[:, 0], points[:, 1], s = 4, c = 'r')
for node in graph:
    node.plot_2D()
plt.show()
```

```python
import time
```

```python
times = []
for i in range(2,7):
    av_time = 0
    for j in range(10):
        points = points = np.random.uniform(-2, 2, (20, i))
        start_time = time.time()
        graph = create_voronoi(points, metric)
        end_time = time.time()
        av_time += end_time - start_time
    times.append(av_time / 10)
```

```python
# plt.plot(list(range(2, 6)), times)
plt.plot(list(range(2, 7)), np.log(times))
```
